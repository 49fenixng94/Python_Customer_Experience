Step 3
- find the optimum number of cluster of KMean
- plot the kmean 

```python
# find optimum number of cluster of KMean

X1=df.loc[:, ['Age', 'Spending Score (1-100)']].values

from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, init='k-means++')
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker = '8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
label = kmeans.fit_predict(X1)
print(label)
```

```python
# check the centroids
## output format = [x, y]

print(kmeans.cluster_centers_)
```

```python
# visualise the centroids
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], color='Black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Apply K-mean to 'Annual Income (k$)', instead of 'Age' 

X2=df.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']].values

from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, init='k-means++')
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker = '8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5)
label = kmeans.fit_predict(X2)
print(label)
```

```python
# check the centroids
## output format = [x, y]

print(kmeans.cluster_centers_)
```

```python
# visualise the centroids
plt.scatter(X2[:,0], X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], color='Black')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Consider all variables into K-mean model

X3 = df.iloc[:, 1:]
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, init='k-means++')
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker = '8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5)
label = kmeans.fit_predict(X3)
print(label)
```

```python
# check the centroids
## output format = [x, y, z]

print(kmeans.cluster_centers_)
```

```python
# visualise the centroids
clusters = kmeans.fit_predict(X3)
df['label'] = clusters

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label == 0], 
           df['Annual Income (k$)'][df.label == 0], 
           df['Spending Score (1-100)'][df.label == 0], c ='blue')
ax.scatter(df.Age[df.label == 1], 
           df['Annual Income (k$)'][df.label == 1], 
           df['Spending Score (1-100)'][df.label == 1], c ='red')
ax.scatter(df.Age[df.label == 2], 
           df['Annual Income (k$)'][df.label == 2], 
           df['Spending Score (1-100)'][df.label == 2], c ='green')
ax.scatter(df.Age[df.label == 3], 
           df['Annual Income (k$)'][df.label == 3], 
           df['Spending Score (1-100)'][df.label == 3], c ='orange')
ax.scatter(df.Age[df.label == 4], 
           df['Annual Income (k$)'][df.label == 4], 
           df['Spending Score (1-100)'][df.label == 4], c ='purple')
ax.view_init(30, 185)

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')

plt.show()
```
